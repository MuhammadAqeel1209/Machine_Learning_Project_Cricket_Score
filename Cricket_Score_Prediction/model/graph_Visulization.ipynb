# Graph Visualization

## Import Libiraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## Load the DataSet

```python
cricket = pd.read_csv('../Data_set/DataSet_Cricket.csv')
```

### Team Performance

```python
team_runs = cricket.groupby('batting_team')['runs'].sum()
team_runs.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Total Runs Scored by Teams')
plt.xlabel('Teams')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

### Run Rate Analysis

```python
cricket['run_rate'] = cricket['runs'] / (cricket['ball'] / 6)
avg_run_rate = cricket.groupby('batting_team')['run_rate'].mean()
avg_run_rate.sort_values(ascending=False).plot(kind='barh', color='orange')
plt.title('Average Run Rate per Over')
plt.xlabel('Run Rate')
plt.ylabel('Teams')
plt.show()
```

### Venue Analysis

```python
# Calculate match counts per venue
venue_counts = cricket['venue'].value_counts()

# Set up the plot
plt.figure(figsize=(12, 8))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette='viridis')

# Add titles and labels
plt.title('Number of Matches Played at Each Venue')
plt.xlabel('Match Count')
plt.ylabel('Venue')

# Adjust font and rotation
plt.xticks(fontsize=8)
plt.yticks(fontsize=5)
plt.tight_layout()

# Show the plot
plt.show()
```
