## Load the DataSet

```python
cricket = pd.read_csv("../Data_set/DataSet_Cricket.csv")
```

```python
# Get the City from Venu Columns of the DataSet and fill the null values of city columns
cricket['city'] = cricket['city'].fillna(cricket['venue'].apply(lambda x : x.split(' ')[0]))
```

```python
# Check all  value of the city
cricket['city'].unique()
```

```python
# Check the count of the cities
cricket['city'].value_counts()
```

```python
eligible_cites = cricket['city'].value_counts()[cricket['city'].value_counts() > 600].index.tolist()
```

```python
eligible_cites
```

```python
# Store the cities column which is eligible cities
cricket = cricket[cricket['city'].isin(eligible_cites)]
```

## Feature Engineering

```python
# Check the score after every ball
cricket['current_score'] = cricket.groupby('match_id')['runs'].cumsum()
```

```python
cricket
```

```python
# Calcluate the balls
cricket['ball_no'] = cricket['ball'].apply(lambda x : str(x).split(".")[1])
```

```python
cricket
```

```python
# Calcluate the overs
cricket['over'] = cricket['ball'].apply(lambda x : str(x).split(".")[0])
```

```python
cricket
```

```python
# Ball Play
cricket['ball_play'] = (cricket['over'].astype(int)*6 + cricket['ball_no'].astype(int))
```

```python
cricket
```

```python
cricket['balls_left'] = 120 -cricket['ball_play']
```

```python
cricket.tail()
```

```python
cricket['balls_left'] = cricket['balls_left'].apply(lambda x:0 if x < 0 else x)
```

```python
cricket.tail()
```

```python
cricket.info()
```

```python
# Player Dismissed fuction for yes or no
cricket['player_dismissed'] = cricket['player_dismissed'].apply(lambda x:1 if x!='0' else '0')
```

```python
cricket.info()
```

```python
cricket['player_dismissed'].value_counts()
```

```python
# Convert to integer
cricket['player_dismissed'] = cricket['player_dismissed'].astype(int)
```

#### Cumulative sum is the running total of values in a column or series, where each value in the result is the sum of all previous values plus the current value.

```python
# Cumulative sum of the `player_dismissed` column for each group of `match_id` in the `cricket` DataFrame
cricket['player_dismissed'] = cricket.groupby('match_id')['player_dismissed'].cumsum()
```

```python
cricket['wicket_left'] = 10 - cricket['player_dismissed']
```

```python
cricket
```

```python
cricket['current_run_rate'] = (cricket['current_score']*6) / cricket['ball_play']
```

```python
groups = cricket.groupby('match_id')
```

```python
match_id = cricket['match_id'].unique()
```

```python
last_five =[]
```

```python
# The code calculates a rolling sum of the 'runs' column over a 30-day window for each group identified by `id` in the `match_id` list and extends the result to `last_five`.
for id in match_id:
    group = groups.get_group(id)  
    rolling_sum = group['runs'].rolling(window=30, min_periods=1).sum()
    last_five.extend(rolling_sum.tolist())
```

```python
cricket['last_five'] = last_five
last_five
```

```python
# The function groups the `cricket` DataFrame by `match_id`, sums the `runs` for each match, and then merges this sum back with the original `cricket` DataFrame based on `match_id`.
final_cricket = cricket.groupby('match_id').sum()['runs'].reset_index().merge(cricket, on='match_id')
```

```python
final_cricket
```

## Final DataSet

```python
# Take the Neccessary Columns from the original dataset
final_cricket = final_cricket[['batting_team', 'bowling_team', 'city', 'current_score', 'balls_left', 'wicket_left','current_run_rate', 'last_five', 'runs_x']]
```

```python
# Drop the Null Rows
final_cricket.dropna(inplace=True)
```

```python
# Check the row of final dataset 
final_cricket.isnull().sum()
```

```python
final_cricket.shape
```

```python
# The function randomly shuffles the rows of the Dataset to create a new random order of the data.
final_cricket = final_cricket.sample(final_cricket.shape[0])
```
